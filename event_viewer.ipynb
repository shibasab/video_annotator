```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
label_dir = "./labels/5-4.txt"
video_dir = ""
events = []

with open(label_dir, mode="r") as f:
    video_dir, events, _ = f.read().split("\n")
```

```python
events = [int(i) for i in events.split(" ")[:-1]]
print(events)
```

```python
cap = cv2.VideoCapture(video_dir)
print(cap.isOpened())
```

```python
def visualize_events(vid_dir, events):
    cap = cv2.VideoCapture(vid_dir)
    
    fig = plt.figure(dpi=2300)
    
    plot_num = 1
    row = 2
    col = 4
    for i, event in enumerate(events):
        cap.set(cv2.CAP_PROP_POS_FRAMES, event)
        ret, img = cap.read()
        
        if ret:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            ax = fig.add_subplot(row, col, plot_num)
            plt.tick_params(bottom=False, 
                            left=False, 
                            labelbottom=False,
                            labelleft=False
                           )
            plt.imshow(img)
            
            plot_num += 1
```

```python
visualize_events(video_dir, events)
```
